```python
def average_isoelectric_point(sequence):
    values = []
    longeur = len(sequence)
    for i in range(longeur):
        try:
            values.append(properties.loc[properties['Code'] == sequence[i]]['pI'].values[0])
        except:
            values.append(0)
    return np.sum(values)/longeur
```

```python
def hydrophobicity_fh(sequence):
    values = []
    longeur = len(sequence)
    for i in range(longeur):
        try:
            values.append(float(properties.loc[properties['Code'] == sequence[i]]['FH'].values[0]))
        except:
            values.append(0)
#     print(values)
    return np.sum(values)
```

```python
def hydrophobicity_eh(sequence):
    values = []
    longeur = len(sequence)
    for i in range(longeur):
        try:
            values.append(float(properties.loc[properties['Code'] == sequence[i]]['EH'].values[0]))
        except:
            values.append(0)
    return np.sum(values)
```
